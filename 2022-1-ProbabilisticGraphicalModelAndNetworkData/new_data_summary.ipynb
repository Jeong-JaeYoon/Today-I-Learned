```python
january = pd.read_csv(path_dir+'\\'+zip[0], encoding='cp949')
january
```

```python
rent_day = january['대여일자'].unique()
rent_day
```

```python
rent_time = january['대여시간'].unique()
rent_time
```

```python
station_id = january['대여소번호'].unique()
print(station_id)
print(len(station_id))
```

```python
january_rentday = []

for i in rent_day:
    for j in range(len(rent_time)):
        january_rentday.append(i)

january_rentday = np.array(january_rentday)
print(january_rentday)
print(len(january_rentday))
print(january_rentday.shape)
```

```python
january_renttime = []

for i in range(len(rent_day)):
    for j in rent_time:
        january_renttime.append(j)

january_renttime = np.array(january_renttime)
print(january_renttime)
print(len(january_renttime))
print(january_renttime.shape)
```

```python
np_time = np.concatenate((january_rentday.reshape(1,-1),january_renttime.reshape(1,-1)), axis=0)
print(len(np_time))
print(np_time.shape)
print(np_time)
```

```python
df_time = pd.DataFrame(np_time.T, columns=['대여일자', '대여일시'])
print(len(df_time))
print(df_time.shape)
df_time
```

```python
station_id = pd.DataFrame(columns=station_id)
station_id
```

```python
df = pd.concat([df_time, station_id], axis=1).fillna(0)
df
```
